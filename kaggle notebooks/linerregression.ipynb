```python
from copy import copy
from sklearn.metrics import mean_squared_error
class LinearRegressionL2:
    
    def __init__(self,alpha= 0.5):
        self.__coef = None
        self._itercept = None
        self.alpha = alpha
    def fit(self,X, y):   
        y_ = y
        y = y.reshape(len(y),1)
        X = self.to_scale_center(X)
        w_0 = -1
        w_x = np.ones(X.shape[1])
        L = 0.0001
        while True:
            w_x_ = copy(w_x)
            w_0_ = w_0
            w_x = w_x + L*(np.sum(X*(y - self.__predict_col(w_0_,w_x_,X) ) , axis = 0) + self.alpha*w_x_)
            w_0 = w_0 + L*(np.sum(y - self.__predict_col(w_0_,w_x_,X)  ) + self.alpha*w_0_)
            mse_before = mean_squared_error(y_,self.__predict_row(w_0_,w_x_,X))
            mse_after = mean_squared_error(y_,self.__predict_row(w_0,w_x,X))
            if mse_before < mse_after:
                w_x = w_x_
                w_0 = w_0_
                L/=10
                continue
            print(abs(mse_before - mse_after))
            if abs(mse_before - mse_after) < 10**(-3):
                break
       
        self.__intercept = w_0
        self.__coef = w_x
        

#Добавить единичный столбец
    def __predict_col(self,w_0,w_x,X):
        p = np.sum(w_0 + w_x*X,axis = 1)
        return p.reshape(p.shape[0],1)
# Маштабирование и цетрирование 
    @staticmethod
    def to_scale_center(X):
        X_scalled = (X - X.mean(axis = 0))/X.std(axis = 0)
        idx = np.isnan(X_scalled)
        X_scalled[idx] = 0
        return X_scalled 
    
    def predict(self,X):
      
        if self.__intercept == None:
            raise ValueErrore
        w_0, w_x = ( self.__intercept, self.__coef)
        X = self.to_scale_center(X)
        y_pred =self.__predict_row(w_0,w_x,X) 
        idx =y_pred<0 
        y_pred[idx] = y_pred.mean()
        return  y_pred
    
    def __predict_row(self,w_0,w_x,X):
        p = np.sum(w_0 + X*w_x,axis = 1)
        return p
    

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 10,test_size = 0.33)
```

```python
model = LinearRegressionL2(alpha=0.5)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
r2 = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmsle_ = mean_squared_log_error(y_test,y_pred)
print(f'mse: {mse} \nr2:{r2}\nrmsle: {rmsle_}')
```
